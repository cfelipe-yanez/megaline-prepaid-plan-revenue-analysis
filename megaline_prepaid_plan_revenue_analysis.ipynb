## Agregar datos por usuario

[Ahora que los datos están limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitará mucho el análisis posterior.]

```python
# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.

df_messages['month'] = df_messages['message_date'].dt.to_period('M')

messages_per_user_month = (df_messages.groupby(['user_id', 'month']).size().reset_index(name='messages_sent'))
messages_per_user_month.head()
```

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month

# Fusionar llamadas y minutos
df_total1_2 = pd.merge(calls_per_user_month, minutes_per_user_month, on=['user_id', 'month'], how='outer')

# Fusionar mensajes
df_total1_2 = pd.merge(df_total1_2, messages_per_user_month, on=['user_id', 'month'], how='outer')

# Fusionar internet
df_total1_2 = pd.merge(df_total1_2, mb_per_user_month, on=['user_id', 'month'], how='outer')

df_total1_2.head(10)
```

```python
# Añade la información de la tarifa
df_final_1_2=0
df_final_1_2 = pd.merge(df_total1_2, df_users[['user_id', 'plan']], on='user_id', how='left')

# Mostrar el resultado final
df_final_1_2.head()
```

<div class="alert alert-block alert-success">

Bien, en esta parte basta con el uso de left ya que la data de consumo de usuarios ya está completa.

```python
# Calcula el ingreso mensual para cada usuario
df_final_1_2 = pd.merge(df_final_1_2, df_plans[['plan_name', 'minutes_included', 'messages_included', 'mb_per_month_included','usd_per_minute', 'usd_per_message', 'usd_per_gb', 'usd_monthly_pay']], 
                        left_on='plan', right_on='plan_name', how='left')

# Calcular el número de minutos, mensajes y MB excedidos
df_final_1_2['excess_minutes'] = df_final_1_2['duration'] - df_final_1_2['minutes_included']
df_final_1_2['excess_messages'] = df_final_1_2['messages_sent'] - df_final_1_2['messages_included']
df_final_1_2['excess_mb'] = df_final_1_2['mb_used'] - df_final_1_2['mb_per_month_included']
df_final_1_2.head()

# si no se exedió ponemos 0
df_final_1_2['excess_minutes'] = df_final_1_2['excess_minutes'].clip(lower=0)
df_final_1_2['excess_messages'] = df_final_1_2['excess_messages'].clip(lower=0)
df_final_1_2['excess_mb'] = df_final_1_2['excess_mb'].clip(lower=0)


# Convertir MB a GB
df_final_1_2['excess_gb'] = df_final_1_2['excess_mb'] / 1024
df_final_1_2['excess_gb'] = np.ceil(df_final_1_2['excess_gb'])


# Calcular el costo por exceso (llamadas, mensajes y datos)
df_final_1_2['extra_cost_minutes'] = df_final_1_2['excess_minutes'] * df_final_1_2['usd_per_minute']
df_final_1_2['extra_cost_messages'] = df_final_1_2['excess_messages'] * df_final_1_2['usd_per_message']
df_final_1_2['extra_cost_data'] = df_final_1_2['excess_gb'] * df_final_1_2['usd_per_gb']

df_final_1_2['total_cost'] = df_final_1_2['usd_monthly_pay'] + df_final_1_2['extra_cost_minutes'] + df_final_1_2['extra_cost_messages'] + df_final_1_2['extra_cost_data']

df_final_1_2.fillna(0, inplace=True)

df_final=df_final_1_2[['user_id', 'month', 'plan','total_cost', 'duration', 'messages_sent', 'mb_used', 
          'excess_minutes', 'excess_messages', 'excess_mb', 'extra_cost_minutes', 
          'extra_cost_messages', 'extra_cost_data']]
df_final.head()
```
